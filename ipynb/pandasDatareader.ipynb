```python
import pandas as pd
import pandas_datareader.data as web
import matplotlib.pyplot as plt
import datetime
```

```python
start_date = "2020-01-1"
end_date = datetime.datetime.now()
```

```python
stocks = ["VOO", "VTI"]
data = web.DataReader(stocks, data_source='yahoo', start=start_date, end=end_date)
print(data)
```

```python
close = data['Close']
ax = close.plot(title='Stock Close Prices 2020')
ax.set_xlabel('Date')
ax.set_ylabel('Close')
ax.grid()
plt.show()
```
